```python
import fastf1
import pandas as pd

# Load a complete session from the 2024 Bahrain Grand Prix Qualifying
session = fastf1.get_session(2024, 'Monzo', 'Q')
session.load()

df_info = pd.DataFrame({
    "field": [
        "session_name",
        "event_name",
        "num_laps_in_results",
        "results_columns",
        "laps_head"
    ],
    "Information": [
        session.name,
        session.event['EventName'],
        session.results.shape[0],
        list(session.results.columns),
        session.laps.head().to_dict()
    ]
})

display(df_info)
```

```python
import requests
import json
import pandas as pd

# Endpoint
url = "https://api.jolpi.ca/ergast/f1/2024/driverstandings/"

# Request
response = requests.get(url)


print("HTTP status code:", response.status_code)

data = json.loads(response.text)


driversstandings = data["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]

records = []
for driver in driversstandings:
    records.append({
        "name": driver["Driver"]["givenName"] + " " + driver["Driver"]["familyName"],
        "nationality": driver["Driver"]["nationality"],
        "points": int(driver["points"])
    })

df = pd.DataFrame(records)


print("Number of records returned:", len(df))

print(df.head(3))

df
```
